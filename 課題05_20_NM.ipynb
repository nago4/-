# **妖怪同士の戦闘と持ち物確認のためのプログラム**



## **概要**

このプログラムは妖怪ウォッチの最初の部分のガチャを回す内容と、バトルの部分の一部機能をプログラムで再現したものです。また、戦闘で妖怪やアイテムをゲットすることができ、それをアイテム確認で見ることができます。また、検索の欄で、妖怪の名前を入れると、その妖怪を持っていれば、その詳細を見ることが可能です。



## **操作方法**

操作方法は、まず、初期設定の欄で、適当な名前と、所持するコインの名前を打ち込みます。名前は、ユーザー名となります。「最初の作業」では、最初に入力してもらったコインの名前を入力してもらいます。そうすれば、色にあった妖怪が仲間になります。別の名前を入力してもいいですが、その場合、使い物にならない妖怪が出てきます（もしかしたらすごくつよいものが出てくるかも）。妖怪をゲットしたら、「自由戦闘」でほかの妖怪と戦うことができます。これを実行する前に、最初にランダムで戦う相手が出てきますが、それを見て、戦うか戦わないかをinputで「はい」か「いいえ」を打ち込むことで選択できます。「はい」と選択すれば、戦闘が始まり、自分で攻撃する妖怪をinputで打ち込むことができます。最初のうちは、あまり強い妖怪に挑戦すると負けることが多いかもしれませんが、もし戦闘を重ねて、仲間を増やしていけば、戦闘できる相手も多くなっていくかもしれません。最後に「現在の所持妖怪・所持アイテム・所持妖怪検索」です。これは、実行することで、所持アイテムと仲間の妖怪、また、inputで持っている妖怪の名前を打ち込むと、その妖怪の詳細を知ることができます。持っていない妖怪は集めていきましょう。
